```python
# Read the file time_thread.txt to get all the time values with threads
times_with_thread = []
file = open("time_thread.txt", "r")
for line in file:
  times_with_thread.append(float(line))
  
print(times_with_thread)
```

```python
# Read the file time_without_thread.txt to get all the time values without threads
times_without_thread = []
file = open("time_without_thread.txt", "r")
for line in file:
  times_without_thread.append(float(line))
print(times_without_thread)
```

```python
import matplotlib.pyplot as plt

# Print graph for time with threads
plt.plot(times_with_thread)
plt.ylabel('Time')
plt.xlabel('Time with threads')
plt.show()
```

```python
# Print graph for time without threads
plt.plot(times_without_thread)
plt.ylabel('Time')
plt.xlabel('Time without threads')
plt.show()
```

```python
# Print both lists in a single graph
plt.plot(times_with_thread, label='Time with threads')
plt.plot(times_without_thread, label='Time without threads')
plt.legend()
plt.ylabel('Time')
```

```python
Average_time_with_threads = sum(times_with_thread)/len(times_with_thread)
Average_time_without_threads = sum(times_without_thread)/len(times_without_thread)
print("Average time with threads: ", Average_time_with_threads)
print("Average time without threads: ", Average_time_without_threads)
```

```python
print("Speedup using threads: ", Average_time_without_threads/Average_time_with_threads)
```

```python
# from re import sub
import subprocess
import time

time_vs_Threads = []
# Read the compiled c file and run it
for i in range(100):
  total_time = 0
  for j in range(5):
    start = time.time()
    subprocess.call(["./a.out",'-a','input1.txt','-b','input2.txt','-t', str(i)])
    end = time.time()
    total_time += end - start
  time_vs_Threads.append(total_time/5)
```

```python
time_vs_Threads
```

```python
# Plot the graph for time vs threads
plt.plot(time_vs_Threads)
plt.title('Time vs Threads')
plt.ylabel('Time')
```
